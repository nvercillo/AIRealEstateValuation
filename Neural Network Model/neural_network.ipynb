```python
import pandas as pd
import json
import ast


filename = "housing_data.csv"
```

```python
#The following code strips out the additional attributes from json dump loocated at the last column
# of the housing_data csv file


col_list = ["AdditionalFields", "Lat","Lon"]

df = pd.read_csv(filename, usecols=col_list)

k = 0

for i, j in df.iterrows():

    if i == 0:
        a_json = json.loads(j["AdditionalFields"])

        data = ast.literal_eval(a_json)
        a = list(data.items())

        additional_df = pd.DataFrame(a)
        additional_df = additional_df.T

    else:

        a_json = json.loads(j["AdditionalFields"])
        data = ast.literal_eval(a_json)
        a = list(data.values())
        new_row = pd.DataFrame(a).T
        additional_df = additional_df.append(new_row)

    k = k + 1
```

```python
#converting y from an object to a numerical type for neural network
Y['sold_price'] = Y['sold_price'].str.replace(',', '').astype(float)


#scaling y using standard scaler
yScaler = StandardScaler()
yScaler.fit(Y)
Y = yScaler.transform(Y)
```

```python
from sklearn.model_selection import train_test_split

#seperating dataset into training versus testing data
Xtr,Xts,Ytr,Yts = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
from tensorflow import keras
import keras_tuner as kt
import tensorflow as tf
from keras_tuner import Hyperband,RandomSearch
import keras_tuner
from tensorflow.keras import layers
import numpy as np

from tensorflow.keras.callbacks import EarlyStopping



#simple neural network to test input and out shape and format

model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=[Xtr.shape[1]]),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.mean_squared_error,
              metrics=['mean_squared_error'])



train_x = np.asarray(Xtr)
train_y = np.asarray(Ytr)
# validation_x = np.asarray(Xts)
# validation_y = np.asarray(Yts)


# model.fit(train_x, train_y, epochs=500)
```

```python
model.fit(train_x, train_y, epochs=500)
```

```python

#Below is the setup for the keras tuner


callback = EarlyStopping(monitor='val_mean_absolute_error', patience=10, verbose=1)



# Creating custom R^2 performance metric for keras models
def r2(y_true,y_pred):
    from keras import backend as K
    SS_res = K.sum(K.square(y_true-y_pred))
    SS_tot = K.sum(K.square(y_true - K.mean(y_true)))
    return (1 - SS_res/(SS_tot + K.epsilon()))

def model_builder(hp):
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=[train_x.shape[1]]))
    for i in range(hp.Int("num_layers", 2, 10)): #number of layers
        model.add(
            layers.Dense(
                units=hp.Int("units_" + str(i), min_value=32, max_value=512, step=32), #number of units
                activation="relu",
            )
        )
    model.add(keras.layers.Dense(1))
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    model.compile(
        optimizer=keras.optimizers.Adam(hp.Choice("learning_rate", [1e-2, 1e-3, 1e-4])),
                loss=keras.losses.mean_absolute_error,
              metrics=['mean_squared_error',r2,keras.losses.mean_absolute_error])
    return model
```

```python
tuner = kt.Hyperband(model_builder,
                     objective=keras_tuner.Objective("val_r2", direction="max"),
                     max_epochs=1000,
                     executions_per_trial=1,
                     overwrite=True,
                     directory="my_dir",
                     project_name="KerasRegressor",
                    
)
```

```python

# here the tuner searches through 1000s of model architectures 
tuner.search(train_x, train_y, epochs=1000, validation_split=0.2,callbacks=[callback])
```

```python
#saving the best model architecture the tuner found

bestHparams=tuner.get_best_hyperparameters(num_trials=1)[0]

models_hyper = tuner.get_best_models(num_models=2)
best_model = models_hyper[0]
score = best_model.evaluate(Xts, Yts, verbose=0)
```

```python
#writing the best model to file
model = tuner.hypermodel.build(bestHparams)
model.fit(X, Y, epochs=50, validation_split=0.2)

model.save("./models/amenity_model")


# def f1(y_true, y_pred):
#     return 1

# model = tf.keras.models.load_model(path_to_model, custom_objects={'r2':r2})
```
